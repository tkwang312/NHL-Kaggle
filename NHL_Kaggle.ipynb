```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import sklearn
from sklearn import linear_model

df = pd.read_csv('gdrive/My Drive/game_skater_stats.csv')
df2 = pd.read_csv('gdrive/My Drive/game_goalie_stats.csv')
```

Create Scatterplot of Time on Ice vs. saves

```python
df.shape
time = df2['timeOnIce'].tolist()
saves = df2['saves'].tolist()
```

```python
g1 = df.loc[df.goals>-1,:]
g1.plot.scatter('shots', 'goals');

g2 = df2.loc[df.shots>-1,:]
g2.plot.scatter('timeOnIce', 'saves');
```

Create Linear Regression model of Time on Ice vs Saves

```python
x = pd.DataFrame(df2['timeOnIce'])
y = pd.DataFrame(df2['saves'])

linear = linear_model.LinearRegression()
linear.fit(x,y)

y_pred = linear.predict(x)
plt.plot(x, y_pred, color='black')

plt.scatter(x, y)
plt.xlabel('timeOnIce')
plt.ylabel('saves')
plt.show()
```

```python
print('the slope of this data is',linear.coef_, 'saves per time')
```

Find R^2 value(how close the data fits the regression line) of the regression line

```python
print('Our single linear model had an R^2 of: %0.3f'%linear.score(x[['timeOnIce']], y)) 
```
